# 🎯 How Language Shapes AI-Generated Political Discourse: A Prototype Analysis in English and Dutch

With the rapid rise of **Generative AI (GenAI)** tools like ChatGPT, these technologies are increasingly integrated into everyday life as a medium for **political inquiry, deliberation, and even companionship**. While much of the academic focus has centered on **algorithmic bias** and the **disinformation potential** of large language models (LLMs), very little is known about how people **actually engage with these tools to discuss political topics**, and how such interactions may influence their political attitudes.

Recent research presents a paradox: GenAI has the potential to **reduce belief in conspiracy theories** and **facilitate dialogue across partisan divides**, yet it may also **reinforce existing opinions** and subtly shape user worldviews without transparency. This experimental project responds to that.

We investigate a underexplored factor:  
> Does the **language** in which a political question is asked shape how GenAI (specifically ChatGPT) frames its answer, emotionally, morally, or civically?

In other words:
- Would a Dutch-speaking user receive a **more neutral, system-trusting answer** to the same political prompt than an English-speaking user?
- Does the model show different degrees of **empathy, encouragement, or skepticism** across languages?
- Could such subtle differences, repeated at scale, affect democratic discourse?


## Experimental Design

To explore this, we designed a **comparative prompt-response experiment** across two languages:

- 7 political prompts were crafted in both **English and Dutch**, based on confirmed corruption-related news events.
- Prompts were categorized into four thematic areas:
  - A – Public trust and systemic framing
  - B – Civic action and motivation
  - C – AI as a political risk
  - D – Cultural approaches to corruption
- We then queried **GPT-4 Omni**, using default settings, and recorded the AI’s responses in both languages.

Each response was analyzed using:
- **TextBlob** for English sentiment (polarity & subjectivity)
- **DutchSentimentAnalysis** for Dutch sentiment scoring and labeling


This pilot study serves as a **computational text analysis precursor** to the broader doctoral research agenda:  
> How do GenAI conversations influence political opinion formation, toward **deliberation** or **polarization**?

By showing that the language of interaction alone may influence how GenAI frames democratic issues, this work raises:
- Does GenAI unintentionally modulate political tone across linguistic-cultural boundaries?
- Could such framing affect trust in institutions, perceived agency, or civic willingness to act?

These findings will contribute to understanding the **opportunities and democratic risks of GenAI**.


## Step 1: Prompt and Answer Dataset

### Data Context
The prompts used in this experiment were **designed based on real corruption news events** reported in [May 2025](https://www.transparency.nl/nieuws/2025/05/corruptieschandaal-bij-navo-3-nederlanders-opgepakt/), including:
- NATO procurement-related arrests in the Netherlands
- Reports of corruption in defense contracts contributing to arms trafficking
- The [misuse of AI](https://www.transparency.nl/nieuws/2025/04/naast-kansen-ook-veel-risicos-bij-ai-op-het-gebied-van-corruptie/) in political disinformation and democratic manipulation

All scenarios assume **confirmed or credible corruption events**, ensuring that the prompts do not speculate but reflect actual issues in public discourse.

### Prompt Construction
Each prompt was carefully formulated in **both English and Dutch**, and fed into **ChatGPT-4.0** to generate a neutral, full-length answer. The prompts were divided into four **thematic categories**:

| Category | Meaning |
|----------|---------|
| **A – Systemic Trust** | How people perceive corruption within institutions (e.g., "Is corruption inevitable?") |
| **B – Civic Action** | Whether citizens feel empowered to act against corruption (e.g., "What can we do?") |
| **C – AI Risk** | Concerns about AI being misused for corruption or disinformation |
| **D – Cultural Framing** | Whether corruption is treated differently across political cultures |

Each category contains two prompts (one in each language) for controlled comparison.

### Output Content
- **Prompt**: User-initiated question (English or Dutch)
- **Answer**: GPT-4 Omni's response in the same language
- All responses were generated using default ChatGPT behavior (no system prompt modification).

The following are shown in English and Dutch respectively.

### Step 2: Sentiment Analysis of English Responses

### Objective:
This step aims to explore how ChatGPT frames political corruption issues **in emotional terms** in its English responses. Specifically, we are interested in whether the answers:

- Adopt a more **positive or negative tone**,
- Are **subjective (opinion-based)** or **objective (fact-based)**,
- Differ **between prompt types** (e.g., "Is corruption inevitable?" vs. "What can citizens do?"),
- Show **inconsistencies across individual prompts** (A1, A3, etc.), even within the same category.

Language framing plays a crucial role in shaping users' perceptions. Previous research (e.g., Lakoff, 2004; Entman, 1993) shows that **how issues are linguistically framed** affects civic attitudes, motivation, and trust.

By applying **sentiment analysis**, we measure:
- The **emotional valence** of ChatGPT's answers (negative vs. positive),
- The **degree of subjectivity** (factual vs. opinionated framing),
These are useful proxies for evaluating **affective framing** and **epistemic positioning**.

### Method:
We use [**TextBlob**](https://textblob.readthedocs.io/en/dev/), a standard Python NLP library, to analyze the sentiment of each answer. For each response, we compute:

- **Polarity**  
  Numerical value between -1.0 and 1.0  
  - **> 0**: positive tone  
  - **< 0**: negative tone  
  - **≈ 0**: neutral or mixed

- **Subjectivity**  
  Numerical value between 0.0 and 1.0  
  - **≈ 0**: more factual, analytical  
  - **≈ 1**: more opinionated, personal

These scores help identify whether ChatGPT:
- Promotes civic engagement (often framed positively),
- Emphasizes helplessness or system failure (often framed negatively),
- Takes a neutral informational stance.

While each prompt (A1, A3, etc.) is unique, they are grouped under broader categories that reflect **thematic intent**. By aggregating polarity and subjectivity within each category, we can:
- Compare overall **framing tendencies** (e.g., “Does ChatGPT frame civic action prompts more optimistically than systemic corruption prompts?”)
- Control for prompt-level noise and interpret **category-level patterns**

**However**, we also keep **individual prompt scores** (e.g., A1, A3) to:
- Identify **outliers or inconsistencies**
- Ensure **micro-level interpretability**

This dual approach (aggregate + individual) balances interpretability and structure.


In the next step, we will visualize:
- Sentiment scores by category (mean polarity and subjectivity),
- Individual scores per prompt (A1, A3, etc.) for micro-level differences.


```python
!pip install textblob
from textblob import TextBlob

# Analyze sentiment for each English response
df_en['Polarity'] = df_en['Answer'].apply(lambda x: TextBlob(x).sentiment.polarity)
df_en['Subjectivity'] = df_en['Answer'].apply(lambda x: TextBlob(x).sentiment.subjectivity)

# Display results for each prompt (hide index)
df_en_display = df_en[['Prompt_ID', 'Category', 'Polarity', 'Subjectivity']]
df_en_display.style.hide(axis='index')

```

## Step 3.1: Interpretation of Sentiment Analysis on Dutch Responses

We now interpret the sentiment scores (`NL_Score`) and labels (`NL_Label`) obtained from the **DutchSentimentAnalysis** package.

The package uses a **Dutch sentiment dictionary**, each assigned a numeric score from **-2 to +2**:
- `+2`: strongly positive
- `+1`: positive
- ` 0`: neutral
- `-1`: negative
- `-2`: strongly negative

The algorithm:
- Matches input text against the dictionary
- Adjusts for **negations** (e.g. "niet goed")
- Averages across all recognized terms in the input
- Outputs a final score (e.g., 0.63 = moderately positive)

Each final score is then mapped to a sentiment label:
- **positive**: score > 0.2
- **neutral**: score between -0.2 and 0.2
- **negative**: score < -0.2

### Result

  - A4: Clearly positive (constructive tone about transparency and reform)

  - B4: Slightly negative, but not strongly enough to be labeled negative

- **Category C (AI risk framing):**
  - Both prompts yield scores close to zero, ChatGPT takes a measured, cautious tone

- **Category D (Cultural comparison):**
  - Extremely neutral, the language is analytical, descriptive, no emotional charge


These scores suggest that:
- ChatGPT in Dutch maintains a **cautious and reserved emotional tone** on corruption-related issues.
- Even positive responses (e.g., B2) stay within moderate emotional bounds.
- No **explicitly negative** tone appears, indicating a diplomatic or institutional alignment in tone.

This might reflect:
- ChatGPT’s design to minimize emotional language in sensitive policy topics
- A possible language calibration difference (Dutch language model is more formal, or training data is more bureaucratic)

## Step 5 Limitations and Future Work

While this experiment provides initial insights into how ChatGPT frames political corruption across languages, several important limitations should be acknowledged:

### 1. Prompt framing bias
All prompts were constructed from a **trusting position toward the corruption allegations**, as they were derived from recent Dutch news reports. As such, the prompts assumed that the corruption claims were real and serious. This framing may have constrained ChatGPT’s possible responses, encouraging alignment rather than allowing for skeptical, defensive, or critical perspectives. Future iterations should include **prompts that express doubt or challenge the corruption narrative**. This would help explore whether GenAI aligns with dominant narratives or balances competing perspectives.

### 2. Dutch sentiment analysis limitations
The Dutch sentiment analysis was conducted using the **DutchSentimentAnalysis package**, which is based on a fixed sentiment lexicon of ~5,000 words. While it accounts for negations and performs well in product review contexts, it may struggle with:
- **Subtle political hedging**, e.g., "niet helemaal transparant" (not fully transparent)
- **Nuanced legal/administrative language**
- **Emotionally neutral but ideologically loaded terms**, such as "verantwoordelijkheid" (responsibility)

In contrast to the more mature tools available for English (e.g.,BERT-based classifiers), the Dutch tool may underdetect emotional polarity or over-rely on lexical features. Future analysis could benefit from **context-aware models or human-coded validation**.

### 3. Lack of situational context
This experiment analyzed **isolated Q&A prompts**, without simulating realistic interactional dynamics. In real-world usage, users bring:
- Personal stakes
- Prior beliefs
- Emotional tone in multi-turn conversations

This context is **crucial to understanding how GenAI might reinforce, challenge, or reshape political beliefs**. Our current analysis does not account for these dynamics. Therefore, results should be interpreted as preliminary insight into **static framing differences**, not dynamic influence.
